```python

X=df_train.drop("Class",inplace=False,axis=1)
Y=df_train["Class"]

transformer1 = ColumnTransformer(transformers=[
    ('simple_impute',SimpleImputer(),['CB','CC','DU','FC','FL','FS','GL']),
    ('Knn_impute',KNNImputer(),['BQ','EL']),
    ('OHE',OneHotEncoder(drop='first'),['EJ'])

],remainder='passthrough')


transformer2 = ColumnTransformer(
    [('log',FunctionTransformer(np.log1p),slice(0, X.shape[1]))]
    ,remainder='passthrough')


transformer3 = ColumnTransformer([
    ('pca', PCA(n_components=40),slice(0, X.shape[1])),
])

Model = Pipeline([
    ('preprocessing', transformer1),
    ('log_transform', transformer2),
    ('pca', transformer3),
    ('logistic_regression', LogisticRegression())
])

Model.fit(X,Y)
```

```python
from sklearn.ensemble import RandomForestClassifier
X=df_train.drop("Class",inplace=False,axis=1)
Y=df_train["Class"]

transformer1 = ColumnTransformer(transformers=[
    ('simple_impute',SimpleImputer(),['CB','CC','DU','FC','FL','FS','GL']),
    ('Knn_impute',KNNImputer(),['BQ','EL']),
    ('OHE',OneHotEncoder(drop='first'),['EJ'])

],remainder='passthrough')


transformer2 = ColumnTransformer(
    [('log',FunctionTransformer(np.log1p),slice(0, X.shape[1]))]
    ,remainder='passthrough')


transformer3 = ColumnTransformer([
    ('pca', PCA(n_components=32),slice(0, X.shape[1])),
])

Model = Pipeline([
    ('preprocessing', transformer1),
    ('log_transform', transformer2),
    ('pca', transformer3),
    ('rfc', RandomForestClassifier( random_state=0))
])

Model.fit(X,Y)
```

```python

Model.predict_proba(X_test[1:])
```

```python

from sklearn.model_selection import GridSearchCV
from sklearn.decomposition import PCA
from sklearn.preprocessing import FunctionTransformer
import numpy as np

n_components_values = [8, 16, 24, 32, 40]

param_grid = {
    'pca__pca__n_components': n_components_values
}

# Create the pipeline with the RandomForestClassifier and GridSearchCV
Model = Pipeline([
    ('preprocessing', transformer1),
    ('log_transform', transformer2),
    ('pca', transformer3),
    ('rfc', RandomForestClassifier(random_state=0))
])


grid_search = GridSearchCV(Model, param_grid, cv=5, scoring='f1_macro')


grid_search.fit(X, Y)


results = grid_search.cv_results_

best_index = np.argmax(results['mean_test_score'])
best_n_components = results['param_pca__pca__n_components'][best_index]
best_mean_test_score = results['mean_test_score'][best_index]

print("Best n_components:", best_n_components)
print("Best mean test score:", best_mean_test_score)
```

```python
from sklearn.ensemble import RandomForestClassifier
X=df_train.drop("Class",inplace=False,axis=1)
Y=df_train["Class"]

transformer1 = ColumnTransformer(transformers=[
    ('simple_impute',SimpleImputer(),['CB','CC','DU','FC','FL','FS','GL']),
    ('Knn_impute',KNNImputer(),['BQ','EL']),
    ('OHE',OneHotEncoder(drop='first'),['EJ'])

],remainder='passthrough')


transformer2 = ColumnTransformer(
    [('log',FunctionTransformer(np.log1p),slice(0, X.shape[1]))]
    ,remainder='passthrough')


transformer3 = ColumnTransformer([
    ('pca', PCA(n_components=8),slice(0, X.shape[1])),
])

Model = Pipeline([
    ('preprocessing', transformer1),
    ('log_transform', transformer2),
    ('pca', transformer3),
    ('rfc', RandomForestClassifier( random_state=0))
])

Model.fit(X,Y)
```

```python

Model.predict_proba(X_test[1:])
```

```python
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(Model, X, Y, cv=5, scoring='accuracy')

mean_cv_score = cv_scores.mean()

print("Cross-validated accuracy score:", mean_cv_score)
```
